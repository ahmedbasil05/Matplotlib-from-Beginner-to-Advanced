**Matplotlib** - helps for data visualization, to find insights and trends

```python
#importing library
import matplotlib.pyplot as plt
```

**Plot** - Line Plot

```python
#plt.plot(x,y, color = 'color', linewidth = val, label = 'label')

months = [1,2,3,4,5]
sales = [1000, 1500, 1200, 1800, 2000]

plt.plot(months, sales, color = 'green', linewidth = 3, label = 'Sales')
```

**Title** - naming whole plot

```python
months = [1,2,3,4,5]
sales = [1000, 1500, 1200, 1800, 2000]

plt.plot(months, sales, color = 'green', linewidth = 3, label = 'Sales')
plt.title('Sales Distribution')
```

**Labels** - helps labelling the axes

```python
months = [1,2,3,4,5]
sales = [1000, 1500, 1200, 1800, 2000]

plt.plot(months, sales, color = 'green', linewidth = 3, label = 'Sales')
plt.title('Sales Distribution')
plt.xlabel('Months')
plt.ylabel('Sales')
```

**Grid**

```python
months = [1,2,3,4,5]
sales = [1000, 1500, 1200, 1800, 2000]

plt.plot(months, sales, color = 'green', linewidth = 3, label = 'Sales')
plt.title('Sales Distribution')
plt.xlabel('Months')
plt.ylabel('Sales')
plt.grid()
```

**Legend**

```python
months = [1,2,3,4,5]
sales = [1000, 1500, 1200, 1800, 2000]

plt.plot(months, sales, color = 'green', linewidth = 3, label = 'Sales')
plt.title('Sales Distribution')
plt.xlabel('Months')
plt.ylabel('Sales')
plt.grid()
plt.legend()
```

**Limits**

```python
months = [1,2,3,4,5]
sales = [1000, 1500, 1200, 1800, 2000]

plt.plot(months, sales, color = 'green', linewidth = 3, label = 'Sales')
plt.title('Sales Distribution')
plt.xlabel('Months')
plt.ylabel('Sales')
plt.grid()
plt.legend()
plt.xlim(0,7)
plt.ylim(700, 2200)
```

**Axes Ticks**

```python
months = [1,2,3,4,5]
sales = [1000, 1500, 1200, 1800, 2000]

plt.plot(months, sales, color = 'green', linewidth = 3, label = 'Sales')
plt.title('Sales Distribution')
plt.xlabel('Months')
plt.ylabel('Sales')
plt.grid()
plt.legend()
plt.xlim(0,7)
plt.ylim(700, 2200)
plt.xticks([1,2,3,4,5])
plt.yticks([1000, 1200, 1400, 1600, 1800, 2000])
```

**Show**

```python
plt.show() #shows the output on the screen
```

**Bars** - comparison between two groups

```python
#plt.bar(x, height, color = 'color', width = value, label = 'label')

product = ['A','B','C','D']
sales = [1000,1500,800,1200]

plt.bar(product, sales, color = 'green')
plt.title('Sales Distribution')
plt.xlabel('Product')
plt.ylabel('Sales')
plt.show()
```

**Pie Charts** - divides data into slices

```python
#plt.pie(x, labels = label_name, autopct='%1.1f%%')

regions = ['North','South','East','West']
sales = [300,1500,500,2000]

plt.pie(sales, labels = regions, autopct='%1.1f%%')
plt.title('Sales Distribution')
plt.show()
```

**Histogram** - frequency distribution

```python
#plt.hist(data, bins = 5, width = 5, color = 'color', edgecolor = 'color')

scores= [85, 92, 76, 88, 95, 67, 74, 81, 90, 78, 84, 69]

plt.hist(scores, bins = 5, width = 5, color = 'yellow', edgecolor = 'orange')
plt.title('Scores Distribution')
plt.xlabel('Scores')
```

**Scatter Plots** - correlation between 2 things

Single Scatter Plot

```python
hours = [3,4,5,6,7,8]
scores = [55,60,65,70,75,80]

plt.scatter(hours, scores, color = 'blue', marker = 'o')
plt.title('Hours vs Scores')
plt.xlabel('Hours')
plt.ylabel('Scores')
plt.show()
```

Multiple Scatter Plot

```python
plt.scatter([1,2,3],[50,60,70], marker='o', color = 'blue')
plt.scatter([1,2,3],[30,45,57], marker='o', color = 'green')
plt.scatter([1,2,3],[70,80,90], marker='o', color = 'red')
plt.show()
```

**Subplots** - multiple plots in same canvas

```python
fig, axes = plt.subplots(2,2, figsize = (10,6))

x = [1,2,3,4,5]
y = [50,60,70,80,90]

axes[0,0].plot(x,y)
axes[0,0].set_title('Line Plot')

axes[0,1].scatter(x,y)
axes[0,1].set_title('Scatter Plot')

axes[1,1].hist(y)
axes[1,1].set_title('Histogram')

axes[1,0].bar(x,y)
axes[1,0].set_title('Bar Plot')

axes[1,0].bar(x,y)
axes[1,0].set_title('Bar Plot')

fig.suptitle('Subplots')
plt.tight_layout
plt.show()
```
